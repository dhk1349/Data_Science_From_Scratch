```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers 
```

```python
x_train=pd.read_csv("PCT\\PCT_train_x.csv")
y_train=pd.read_csv("PCT\\PCT_train_y.csv")
x_test=pd.read_csv("PCT\\PCT_test_x.csv")
y_test=pd.read_csv("PCT\\PCT_test_y.csv")
```

```python
x_train=x_train.drop(columns=["T_ID", "YEAR"])
x_test=x_test.drop(columns=["T_ID", "YEAR"])
```

```python
x_train['target']=y_train['PCT']
```

```python
corr_x_train=x_train.corr()
```

```python
pd.set_option('display.max_columns', None)
```

```python
sign=[0 if abs(e)<0.1 else np.sign(e) for e in corr_x_train['target'].to_list()]
sign=sign[:-1]
```

```python
remove_idx=[idx for idx,i in enumerate(sign) if (i==0)]
```

```python
x_train=x_train.drop(columns=['target'])
```

```python
x_train=x_train.drop(columns=[i for idx,i in enumerate(x_train.columns) if idx in remove_idx])
```

```python
x_test=x_test.drop(columns=[i for idx,i in enumerate(x_test.columns) if idx in remove_idx])
```

```python
sign=[i for i in sign if i!=0]
```

```python
x_train=x_train*sign
```

```python
x_test=x_test*sign
```

```python
y_train=y_train.drop(columns=["T_ID", "YEAR"])
```

```python
def build_model():
  model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[len(x_train.keys())]),
    layers.Dense(128, activation='relu'),
    layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.RMSprop(0.001)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
  return model
```

```python
model = build_model()
model.summary()
```

```python
#    (.)     
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')

EPOCHS = 500

history = model.fit(
  x_train, y_train,
  epochs=EPOCHS, validation_split = 0.2, verbose=2,
  callbacks=[PrintDot()])
```

```python
import matplotlib.pyplot as plt

def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure(figsize=(8,12))

  plt.subplot(2,1,1)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
  plt.ylim([0,0.15])
  plt.legend()

  plt.subplot(2,1,2)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
  plt.ylim([0,0.01])
  plt.legend()
  plt.show()

plot_history(history)
```
