```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data=[1, 2, 3]
y_data=[2, 4, 6]
```

```python
plt.plot(x_data, y_data)
plt.xlabel('Hours')
plt.ylabel('Points')
plt.show()
```

```python
w=1.0
def forward(x):
    return w*x

def loss(x, y):
    y_pred=forward(x)
    return (y_pred-y)**2
```

대충 w를 0부터 4까지 주면서 각 w당 sample의 데이터 예측치, MSE를 보여준다. 

```python
for w in np.arange(0.0, 4.1, 0.1):
    print('w= ', w)
    l_sum=0
    for x_val, y_val in zip(x_data, y_data): #iterate through whole dataset
        y_pred_val=forward(x_val)
        l=loss(x_val, y_val)
        l_sum+=l
        print("\t", x_val, y_val, y_pred_val, l)
        
    print("MSE=", l_sum/3)
```

```python
w_list=[]
mse_list=[]

for w in np.arange(0.0, 4.1, 0.1):
    #print('w= ', w)
    l_sum=0
    for x_val, y_val in zip(x_data, y_data): #iterate through whole dataset
        y_pred_val=forward(x_val)
        l=loss(x_val, y_val)
        l_sum+=l
        #print("\t", x_val, y_val, y_pred_val, l)
    
    w_list.append(w)
    mse_list.append(l_sum/3)
    #print("MSE=", l_sum/3)

plt.plot(w_list, mse_list)
plt.xlabel("w")
plt.ylabel("Loss")
plt.show()
```
