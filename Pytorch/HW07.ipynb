<h1>ex6-1 Try other activation layer<h1>

```python
import torch
from torch.autograd  import Variable
from torch import nn
from torch import sigmoid
import torch.nn.functional as F
import torch.optim as optim
```

```python
x_data=Variable(torch.Tensor([[1.0],[2.0],[3.0],[4.0]]))
y_data=Variable(torch.Tensor([[0.],[0.],[1.],[1.]]))
```

```python
class Model(torch.nn.Module):
    def __init__(self, activation):
        super(Model, self).__init__()
        self.linear=torch.nn.Linear(1,1)
        self.A=activation
        if(activation=='ReLU'):
            self.activation=F.relu
        elif(activation=='ReLU6'):
            self.activation=F.relu6
        elif(activation=='ELU'):
            self.activation=F.elu
        elif(activation=='SELU'):
            self.activation=F.selu
        elif(activation=='PReLU'):
            self.activation=F.prelu
        elif(activation=='LeakyReLU'):
            self.activation=F.leaky_relu
        elif(activation=='Threshold'):
            self.activation=nn.Threshold(0.5, 0)
        elif(activation=='Hardtanh'):
            self.activation=F.hardtanh
        elif(activation=='Sigmoid'):
            self.activation=F.sigmoid
        elif(activation=='Tanh'):
            self.activation=F.tanh
    def forward(self,x):
        if(self.A=="Threshold"):
            y_pred=self.activation(self.linear(x))

        elif(self.A=="PReLU"):
            y_pred=self.activation(self.linear(x), torch.tensor(0.25))
        else:
            y_pred=self.activation(self.linear(x))
        return y_pred
```

```python
activationlst=['ReLU','ReLU6','ELU', 'SELU', 'PReLU', 'LeakyReLU', 'Threshold', 'Hardtanh', 'Sigmoid', 'Tanh']
```

```python
model_dict={}
```

```python
for A in activationlst:
    model_dict[A]=Model(A)
```

```python
model_dict
```

```python
criterion=torch.nn.MSELoss(reduction='mean')
```

```python
for A in activationlst:
    optimizer=torch.optim.SGD(model_dict[A].parameters(), lr=0.01)
    print(f"Activation: {A}")
    for epoch in range(1000):
        y_pred=model_dict[A](x_data)
        
        loss=criterion(y_pred, y_data)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if(epoch%100==0):
            print(f'EPOCH: {epoch} | LOSS: {loss.item():.4f}')
        
    hour_var=torch.tensor([[1.0]])
    print("predict 1 hour ", 1.0, model_dict[A](hour_var).item()>0.5)
    hour_var=torch.tensor([[7.0]])
    print("predict 7 hour ", 1.0, model_dict[A](hour_var).item()>0.5, "\n")
```

<h1>ex6-1 Try different optimizer<h1>

```python
optimizer_dict={'Adadelta':optim.Adadelta, 'Adagrad':torch.optim.Adagrad, 'Adam':torch.optim.Adam, 'Adamax':torch.optim.Adamax, 'ASGD':torch.optim.ASGD,
'RMSprop':torch.optim.RMSprop, 'Rprop':torch.optim.Rprop, 'SGD':torch.optim.SGD}
```

```python
optimizerlst=['Adadelta', 'Adagrad', 'Adam', 'Adamax', 'ASGD', 'RMSprop', 'Rprop', 'SGD']
for opt in optimizerlst:
    model2=Model('Sigmoid')
    criterion=torch.nn.BCELoss(reduction='mean')
    
    print(f"optimizer: {opt}")
    if(opt=='SGD'):
        optimizer=optimizer_dict[opt](model2.parameters(), 0.01)
    else:
        optimizer=optimizer_dict[opt](model2.parameters())
    for epoch in range(100):
        y_pred=model2(x_data)
        
        loss=criterion(y_pred, y_data)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if((epoch+1)%10==0):
            print(f'EPOCH: {epoch+1} | LOSS: {loss.item():.4f}')
        
    hour_var=torch.tensor([[1.0]])
    print("predict 1 hour ", 1.0, model2(hour_var).item()>0.5)
    hour_var=torch.tensor([[7.0]])
    print("predict 7 hour ", 1.0, model2(hour_var).item()>0.5, "\n")
```
