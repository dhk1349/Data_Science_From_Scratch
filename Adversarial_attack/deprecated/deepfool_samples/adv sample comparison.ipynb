```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
total=np.load("./total.npy", allow_pickle=True)
normal=np.load("./normal_samples.npy", allow_pickle=True)
perturbed=np.load("./perturbed_samples.npy", allow_pickle=True)
```

```python
plt.rcParams["figure.figsize"] = (200, 200)
```

```python
cnt=0
for i in range(len(total)):
    cnt += 1
    plt.subplot(10, 10, cnt)
    plt.xticks([], [])
    plt.yticks([], [])

    orig, adv, _perturbed, _  = total[i]
    plt.title("{} -> {}".format(orig, adv))
    plt.imshow(_perturbed, cmap="gray")
plt.tight_layout()
plt.show()
```

```python
cnt=0
for i in range(len(total)):
    cnt += 1
    plt.subplot(10, 10, cnt)
    plt.xticks([], [])
    plt.yticks([], [])

    orig, adv, _, _normal  = total[i]
    plt.title("{} -> {}".format(orig, adv))
    plt.imshow(_normal, cmap="gray")
```

```python
cnt=0
for i in range(len(total)):
    cnt+=1
    if(cnt%10!=0):
        orig, adv, _, _  = total[i]
        print(f"{orig}->{adv}", end=" ")
    else:
        print()
```
